```python

#! Lab1: Analyze the relationship between the size of houses (measured in square footage) and their selling prices in a particular neighborhood. You have collected data on various houses in that neighborhood.Create a scatter plot using the below data and share your conclusion/analysis

import numpy as np
import matplotlib.pyplot as plt
# Input:
square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800, 3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])

plt.scatter(square_footage, selling_prices, color="g")
plt.xlabel("Square Footage (sq. ft.)")
plt.ylabel("Selling Price ($000s)")
plt.title("House Pricing vs Square Footage")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

#! Lab4: Suppose you're a sales manager for an e-commerce company, and you want to create a figure with subplots to compare the sales performance of different product categories over time. You have sales data for four product categories: Electronics, Clothing, Home & Garden, and Sports & Outdoors. Share your conclusion/analysis.

import matplotlib.pyplot as plt

# Input:
months = np.arange(1, 13)
electronics_sales     = np.array([25000, 28000, 31000, 27000, 30000, 32000, 
                                  35000, 36000, 38000, 39000, 41000, 42000])
clothing_sales        = np.array([15000, 16000, 17000, 18000, 19000, 20000, 
                                  21000, 22000, 23000, 24000, 25000, 26000])
home_garden_sales     = np.array([18000, 19000, 20000, 21000, 22000, 23000, 
                                  24000, 25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 
                                  18000, 19000, 20000, 21000, 22000, 23000])

plt.subplot(2, 2, 1)
plt.plot(months, electronics_sales, marker=".", markersize=8, color='b')
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.title("Electronics")

plt.subplot(2,2,2) 
plt.plot(months, clothing_sales, marker=".", markersize=8, color='g')
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.title("Clothing")

plt.subplot(2,2,3) 
plt.plot(months, home_garden_sales, marker=".", markersize=8, color='r')
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.title("Home and Garden")

plt.subplot(2,2,4)
plt.plot(months, sports_outdoors_sales, marker=".", markersize=8, color='purple')
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.title("Sports and Outdoor")
plt.tight_layout()
plt.show()
```

```python

#* CHATGPT EXERCISE

#! Scenario:Analyzing Sales Data Suppose you work for a retail company, and you have dummy data containing sales data for the past year. The data includes information such as SalesDate,product names,regions, sales quantities, prices, and dates. You have to generate a bar chart ,pie plot on region and prices and line chart on SalesDate and prices columns. Further, you need to get some inference out of the chart. Create a ChatGPT prompt to generate the code for this scenario. Based on the code generated, ask ChatGPT to give the conclusion/inference. Note. You can provide the data to ChatGPT or ask it to use sample data.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = {
    'SalesDate': pd.date_range(start='2023-01-01', periods=100, freq='D'),
    'ProductName': np.random.choice(['Product A', 'Product B', 'Product C'], 100),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 100),
    'SalesQuantity': np.random.randint(1, 20, 100),
    'Price': np.random.uniform(10, 100, 100)
}

df = pd.DataFrame(data)

# Bar chart: Total Sales Quantity by Region
sales_by_region = df.groupby('Region')['SalesQuantity'].sum()
plt.figure(figsize=(10, 6))
sales_by_region.plot(kind='bar', color='skyblue')
plt.title('Total Sales Quantity by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales Quantity')
plt.grid(axis='y')
plt.show()

# Pie chart: Distribution of Total Sales by Region
plt.figure(figsize=(8, 8))
sales_by_region.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Distribution of Total Sales by Region')
plt.ylabel('')
plt.show()

# Line chart: Trend of Average Prices Over Time
df['SalesDate'] = pd.to_datetime(df['SalesDate'])
avg_price_by_date = df.groupby('SalesDate')['Price'].mean()
plt.figure(figsize=(12, 6))
avg_price_by_date.plot(kind='line', marker='o', color='green')
plt.title('Trend of Average Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```
