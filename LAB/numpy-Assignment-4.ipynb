#### 1. Calculate the total revenue generated by two product categories in a store<hr/>
Input: <br/>
`category1 revenue np.array([500, 600, 700, 550)`<br/>
`category2 revenue np.array([450, 700, 800, 600) `<br/>
Output: <br/>
`Total Revenue: 950 1300 1500 1150] `

```python
import numpy as np  

category1_revenue = np.array([500, 600, 700, 550])
category2_revenue = np.array([450, 700, 800, 600]) 

print("Total Revenue: ", category1_revenue + category2_revenue)
```

#### 2. Calculate the profit made by a company <hr/>
Input:<br/>
`revenue  = np.array([10000, 12000, 11000, 10500)` <br/>
`expenses = np.array([4000, 5000, 4500, 4800)`

Output:<br/>
`Profit: [6000 7000 6500 5700]`


```python
import numpy as np

revenue  = np.array([10000, 12000, 11000, 10500])
expenses = np.array([4000, 5000, 4500, 4800])

profit = revenue - expenses

print("Profit: ", profit)
```

#### 3. Determine which products in a store are out of stock (quantity is 0).<hr/>
Input: <br/>
`inventory= np.array([10, 0, 5, 0, 20, 0])`<br/>

Output: <br/>
`Out of Stock Products: [0 0 0]`

```python
import numpy as np
inventory = np.array([10, 0, 5, 0, 20, 0])
outOfStock = np.where(inventory == 0)[0]

print("Out of Stock Products: ", outOfStock)
```

#### 4. Calculate the total cost of items in a shopping cart, considering the quantity and price<hr/>

Input: <br/>
`quantity = np.array([2, 3, 4, 1])`<br/>
`price_per_items = np.array([10.0, 5.0, 8.0, 12.0])`<br/>

Output: <br/>
`Total cost of items : [20. 15. 32. 12.]`

```python
import numpy as np
quantity = np.array([2, 3, 4, 1])
price_per_items = np.array([10.0, 5.0, 8.0, 12.0])

print("Total cost of items : ", quantity * price_per_items)
```
